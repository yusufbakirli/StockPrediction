```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Örneğin Apple (AAPL) hisse senedi için veri çekme
data = yf.download("AAPL", start="2020-01-01", end="2023-12-01")

# Verilerin ilk 5 satırını görüntüleme
print(data.head())
```

```python
data['Close'].plot(title="Hisse Senedi Kapanış Fiyatı")
plt.show()
```

```python
data.isnull().sum()
data.dropna(inplace=True)
```

```python
for i in range(1, 11):
    data[f"Close_lag_{i}"] = data['Close'].shift(i)
data.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
```

```python
X = data[[f"Close_lag_{i}" for i in range(1, 11)]].values
y = data['Close'].values
```

```python
# Veriyi ölçeklendirme
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Veriyi LSTM için 3 boyuta dönüştürme
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
# Modelin tanımlanması
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    tf.keras.layers.LSTM(50),
    tf.keras.layers.Dense(1)
])
```

```python
# Modelin derlenmesi
model.compile(optimizer='adam', loss='mse')
```

```python
# Modelin eğitilmesi
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_test, label="Gerçek Değerler")
plt.plot(predictions, label="Tahminler")
plt.legend()
plt.show()
```
